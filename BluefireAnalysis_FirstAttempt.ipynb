```python
def process_county(county_data):
    county_data = county_data.sort_values('RecordDateTime')
    outages = []
    current_outage = {}
    
    for i, row in county_data.iterrows():
        if row['CustomersOut'] > 0:
            if not current_outage:
                current_outage = {
                    'start_time': row['RecordDateTime'],
                    'end_time': None,
                    'customers_affected': [],
                    'county': row['CountyName']
                }
            current_outage['customers_affected'].append(row['CustomersOut'])
        else:
            if current_outage:
                current_outage['end_time'] = row['RecordDateTime']
                outages.append(current_outage)
                current_outage = {}
    
    if current_outage and 'end_time' not in current_outage:
        current_outage['end_time'] = county_data['RecordDateTime'].iloc[-1]  # or set to None
        outages.append(current_outage)
    
    results = []
    for outage in outages:
        duration = outage['end_time'] - outage['start_time']
        mean_customers_affected = sum(outage['customers_affected']) / len(outage['customers_affected'])
        results.append({
            'UtilityName': county_data['UtilityName'].iloc[0],
            'StateName': county_data['StateName'].iloc[0],
            'CountyName': outage['county'],
            'start_time': outage['start_time'],
            'end_time': outage['end_time'],
            'customer_affected_mean': mean_customers_affected,
            'duration': duration
        })
    return pd.DataFrame(results)

ztz_results = pd.DataFrame()
for county, group in data_july.groupby(['CountyFIPS', 'CountyName']):
    county_results = process_county(group)
    ztz_results = pd.concat([ztz_results, county_results], ignore_index=True)

print(ztz_results.head())
```

# Change Point Detection with the PELT (Pruned Exact Linear Time) Algorithm

(Reference paper: https://www.sciencedirect.com/science/article/pii/S0165168419303494?via%3Dihub)

### How PELT Works

1. **Segmentation**:  
   The algorithm divides the time series into segments where the properties (mean, variance, etc.) are constant within each segment but change at the boundaries (change points).
   

2. **Cost Function**:  
   The cost function, \( C(t) \), is defined for each segment \( t \) and is typically based on the chosen model (e.g., sum of squared deviations for \( l2 \)).

   $$
   C(t) = \sum_{i=t_1}^{t_2} (y_i - \hat{y})^2
   $$

   for a segment \([$t_1$, $t_2$]\) in the \(l2\) model, where \($y_i$) is the data point and \($\hat{y}$) is the mean of the segment.

3. **Penalty Term**:  
   The total cost includes a penalty term \($\beta$\) for each change point to control the complexity of the model.

   $$
   \text{Total Cost} = \sum_{\text{segments}} C(t) + \beta \times \text{(number of change points)}
   $$

   The penalty term \($\beta$\) helps balance between fitting the data well and avoiding too many change points (overfitting).

### Penalty Impact on Change Points

- **High Penalty (\($\beta$\))**:  
  Results in fewer change points, potentially missing some actual changes but reducing noise.  
  Example: For a high \($\beta$\), only major changes like from 10 to 50 are detected, ignoring smaller fluctuations.

- **Low Penalty (\($\beta$\))**:  
  Allows more change points, capturing more details in the data but possibly introducing noise.  
  Example: For a low \($\beta$\), detects changes even for small shifts like from 10 to 12, 12 to 14, etc.

### Numerical Example

Consider a time series: [1, 1, 2, 2, 10, 10, 50, 50, 2, 2, 1, 1]

- **High Penalty**:  
  Detects fewer change points: [1, 1, 2, 2, **10**, 10, **50**, 50, 2, 2, 1, 1]  
  Change points detected at 10 and 50.

- **Low Penalty**:  
  Detects more change points: [1, 1, **2**, 2, **10**, 10, **50**, 50, **2**, 2, **1**, 1]  
  Change points detected at 2, 10, 50, 2, and 1.


### Rules of application: 
1. Zero always terminates a previous outage, regardless of any other conditions.
2. After a zero, any non-zero value starts a new outage.
3. If the segment count is less than or equal to 2, it always forms a new group.
4. Decreasing values should not be counted as new outage.
5. A new outage can only be created when the value increases from the previous value. However, the final decision on whether to start a new group should be determined by the PELT algorithm (even if the value increases from the previous one, it might still belong to the same group).

### Models Explored:
1. l1 - Based on the L1 norm (absolute distance). Example: The distance between 1 and 2 is |1 - 2| = 1.
2. l2 - Based on the L2 norm (Euclidean distance). Example: The distance between 1 and 2 is sqrt((1 - 2)^2) = 1.
3. rbf - Utilizes Radial Basis Functions (RBF) to capture non-linear changes. Utilizes the distance from a central point and can model data. Example: [1, 1.5, 2, 2.5, 5, 5.5, 6, 6.5] >> [1, 1.5, 2, 2.5], [5, 5.5, 6, 6.5] >> distance from central point of 1.75 and 5.75 for each segment.
4. linear - Detects changes in the linear trend of the signal. Example: Identifies changes in slope, such as a switch from an increasing to a decreasing trend.
5. normal - Assumes the data follows a normal distribution, detecting changes in both mean and variance. Example: Useful when data has varying averages and spreads.
6. rank - A non-parametric method that detects changes based on ranks. Example: raw data - [4, 3, 2, 1, 10, 9, 8, 7] >> rank data - [4, 3, 2, 1, 8, 7, 6, 5] >> change point at position 4
7. ar - Detects changes in autoregressive (AR) models. Example: Useful for time series data where current values depend on previous values. (failed to apply due to the short-period-nature of the data) 


```python
def detect_outages(data, model, penalty):
    if data.empty:
        return pd.DataFrame(), []

    data = data.reset_index(drop=True)
    data['CustomersOut'] = pd.to_numeric(data['CustomersOut'], errors='coerce').fillna(0)
    data['CustomersTracked'] = pd.to_numeric(data['CustomersTracked'], errors='coerce').fillna(1)

    segments = []
    current_segment_start = 0
    last_non_zero = None
    prev_zero_TF = False

    for i in range(len(data)):
        current_value = data['CustomersOut'][i]
        if current_value == 0:
            if last_non_zero is not None and prev_zero_TF == False:
                segments.append((current_segment_start, i + 1))
            if prev_zero_TF == False:
                current_segment_start = i + 1
            last_non_zero = None
            prev_zero_TF = True
        else:
            if last_non_zero is None:
                current_segment_start = i
            last_non_zero = current_value
            prev_zero_TF = False
            
    if last_non_zero is not None and current_segment_start < len(data):
        segments.append((current_segment_start, len(data)))

    all_segments = []
    outage_segments = []
    for start, end in segments:
        segment_data = data.iloc[start:end]
        for j in range(start + 1, end):
            if segment_data.at[j, 'CustomersOut'] < segment_data.at[j - 1, 'CustomersOut']:
                segment_data.at[j, 'CustomersOut'] = segment_data.at[j - 1, 'CustomersOut']
        if end - start > 2:
            segment_data['OutageRatio'] = segment_data['CustomersOut'] / segment_data['CustomersTracked']
            points = segment_data['OutageRatio'].values
            algo = rpt.Pelt(model=model, min_size=1, jump=1).fit(points)
            try:
                change_points = algo.predict(pen=penalty)
                actual_change_points = [start + cp for cp in change_points if cp != 0]  # Ensure no zero index changes
                all_segments.extend(actual_change_points)
                start = actual_change_points[-1] if actual_change_points else start
                if start < end:
                    all_segments.append(start)
            except Exception as e:
                print(f"Segmentation failed: {str(e)}")
                all_segments.append(start)
        else:
            all_segments.append(end)

    for start in all_segments:
        if start < len(data):  
            end = all_segments[all_segments.index(start) + 1] if all_segments.index(start) + 1 < len(all_segments) else len(data)
            segment = data.iloc[start:end]
            if not segment.empty:
                outage_segments.append({
                    'CountyFIPS': segment['CountyFIPS'].iloc[0],
                    'CityName': segment['CityName'].iloc[0],
                    'start_time': segment['RecordDateTime'].iloc[0],
                    'end_time': segment['RecordDateTime'].iloc[-1],
                    'average_customers_out': segment['CustomersOut'].mean(),
                    'duration': segment['RecordDateTime'].iloc[-1] - segment['RecordDateTime'].iloc[0],
                    'weighted_customers_out': segment['CustomersOut'].mean() * (segment['RecordDateTime'].iloc[-1] - segment['RecordDateTime'].iloc[0]).total_seconds(),
                    'segment_label': f"Outage_{start}"
                })

    return pd.DataFrame(outage_segments), all_segments

def evaluate_model(data, mod, pen_val):
    grouped_data = data.reset_index().groupby(['CountyFIPS', 'CityName'])
    results_by_county = grouped_data.apply(lambda x: detect_outages(x, mod, pen_val)[0]).reset_index(drop=True)
    change_points_by_county = grouped_data.apply(lambda x: detect_outages(x, mod, pen_val)[1])
    return results_by_county, change_points_by_county

def binary_search_best_penalty(data, mod, pen_min, pen_max, tolerance):
    global best_frequency_diff, best_fre_mod, best_fre_pen, best_fre_change_points
    
    while pen_max - pen_min > tolerance:
        pen_val = (pen_min + pen_max) / 2
        
        try:
            results_by_county, change_points_by_county = evaluate_model(data, mod, pen_val)
            frequency_diff = abs(len(results_by_county) - answer_frequency)
            print(frequency_diff, mod, pen_val )

            if frequency_diff < best_frequency_diff:
                best_frequency_diff = frequency_diff
                best_fre_mod = mod
                best_fre_pen = pen_val
                best_fre_change_points = change_points_by_county.to_dict()

            if len(results_by_county) > answer_frequency:
                pen_min = pen_val
            else:
                pen_max = pen_val 

        except Exception as e:
            print(f"Error processing model {mod} with penalty {pen_val}: {e}")
            break

            
            
data_july = data_july.sort_values(by=['CountyFIPS', 'CityName', 'RecordDateTime'])

models = ["l2", "rbf", "linear", "l1", "normal", "rank", "ar"]

pen_min = 0.00001
pen_max = 100
tolerance = 0.00001  

best_frequency_diff = 999999
best_fre_mod = "l1"
best_fre_pen = 0.1
best_fre_change_points = {}

for mod in models:
    binary_search_best_penalty(data_july, mod, pen_min, pen_max, tolerance)
```

```python


# def detect_outages(data, model, penalty):
#     points = data['CustomersOut'].values

#     model = model
#     algo = rpt.Pelt(model=model, min_size=1, jump=1).fit(points)

#     try:
#         change_points = algo.predict(pen=penalty)
#     except Exception as e:
#         print(f"Segmentation failed for {data['CountyFIPS'].iloc[0]}: {str(e)}")
#         return pd.DataFrame(), []

#     outage_segments = []
#     for start, end in zip([0] + change_points, change_points + [len(points)]):
#         if end > start: 
#             segment = data.iloc[start:end]
#             duration = segment['RecordDateTime'].iloc[-1] - segment['RecordDateTime'].iloc[0]
#             average_customers_out = segment['CustomersOut'].mean() if 'CustomersOut' in segment.columns else 0
#             outage_segments.append({
#                     'CountyFIPS': data['CountyFIPS'].iloc[0],  
#                     'start_time': segment['RecordDateTime'].iloc[0],
#                     'end_time': segment['RecordDateTime'].iloc[-1],
#                     'average_customers_out': average_customers_out,
#                     'duration': duration,
#                     'weighted_customers_out': average_customers_out * duration.total_seconds()            
#             })

#     return pd.DataFrame(outage_segments), change_points


# def detect_outages(data, model, penalty):
#     # Create a copy of data to manipulate and examine
#     temp_data = data.copy()
#     # Replace 0 with NaN to ignore temporary drops during change point detection
#     temp_data['CustomersOut'].replace(0, pd.NA, inplace=True)

#     # Forward fill to bridge gaps of zeros - treat them as part of the ongoing outage
#     temp_data['CustomersOut'].fillna(method='ffill', inplace=True)

#     # Detect change points
#     points = temp_data['CustomersOut'].dropna().values
#     algo = rpt.Pelt(model=model, min_size=1, jump=1).fit(points)
#     try:
#         change_points = algo.predict(pen=penalty)
#     except Exception as e:
#         print(f"Segmentation failed for {data['CountyFIPS'].iloc[0]}: {str(e)}")
#         return pd.DataFrame(), []

#     # Post-process to integrate change points back into original data
#     outage_segments = []
#     last_end = 0
#     for start, end in zip([0] + change_points, change_points + [len(points)]):
#         if end > start:
#             segment = data.iloc[last_end:end]  # use last_end to end, to cover the full range including zeros
#             if not segment['CustomersOut'].equals(segment['CustomersOut'].shift()):  # Check if all values are the same
#                 duration = segment['RecordDateTime'].iloc[-1] - segment['RecordDateTime'].iloc[0]
#                 average_customers_out = segment['CustomersOut'].mean()
#                 outage_segments.append({
#                     'CountyFIPS': data['CountyFIPS'].iloc[0],
#                     'start_time': segment['RecordDateTime'].iloc[0],
#                     'end_time': segment['RecordDateTime'].iloc[-1],
#                     'average_customers_out': average_customers_out,
#                     'duration': duration,
#                     'weighted_customers_out': average_customers_out * duration.total_seconds()
#                 })
#             last_end = end  # update last_end to the current end

#     return pd.DataFrame(outage_segments), change_points

def detect_outages(data, model, penalty):
    if data.empty:
        return pd.DataFrame(), []

    # Reset index to ensure sequential integer indexing
    temp_data = data.reset_index(drop=True)

    # Preprocess data to handle any decreases as forward fills
    for i in range(1, len(temp_data)):
        if temp_data.loc[i, 'CustomersOut'] < temp_data.loc[i - 1, 'CustomersOut']:
            temp_data.loc[i, 'CustomersOut'] = temp_data.loc[i - 1, 'CustomersOut']

    # Fit the change point detection model
    points = temp_data['CustomersOut'].values
    algo = rpt.Pelt(model=model, min_size=1, jump=1).fit(points)

    try:
        change_points = algo.predict(pen=penalty)
    except Exception as e:
        print(f"Segmentation failed for {temp_data['CountyFIPS'].iloc[0]}: {str(e)}")
        return pd.DataFrame(), []

    # Generate the outage segments
    outage_segments = []
    for start, end in zip([0] + change_points, change_points + [len(points)]):
        segment = data.iloc[start:end]
        if not segment.empty:
            duration = segment['RecordDateTime'].iloc[-1] - segment['RecordDateTime'].iloc[0]
            average_customers_out = segment['CustomersOut'].mean()
            outage_segments.append({
                'CountyFIPS': segment['CountyFIPS'].iloc[0],
                'start_time': segment['RecordDateTime'].iloc[0],
                'end_time': segment['RecordDateTime'].iloc[-1],
                'average_customers_out': average_customers_out,
                'duration': duration,
                'weighted_customers_out': average_customers_out * duration.total_seconds()
            })

    return pd.DataFrame(outage_segments), change_points


data_july = data_july.sort_values(by=['CountyFIPS', 'RecordDateTime'])

# Initialize variables for best models and penalties
models = ["l1", "l2", "rbf"]
models = ["l1", "l2"]
pen_vals = [0.1, 0.5, 1, 2, 3, 5, 7, 10, 15, 20, 30]
pen_vals = [0.1, 1,  5,20]
best_avg_duration_diff = pd.Timedelta(days=9999)
best_frequency_diff = 999999
best_dur_mod = "l1"
best_dur_pen = 0.1
best_fre_mod = "l1"
best_fre_pen = 0.1
best_dur_change_points = {}
best_fre_change_points = {}

# Find the best model and penalty
for mod in models:
    for pen_val in pen_vals:
        try:
            grouped_data = data_july.reset_index().groupby('CountyFIPS')
            results_by_county = grouped_data.apply(lambda x: detect_outages(x, mod, pen_val)[0]).reset_index(drop=True)
            change_points_by_county = grouped_data.apply(lambda x: detect_outages(x, mod, pen_val)[1])

            print("\n\n\nProperty for models: modelname-", mod, " penalty value-", pen_val, "\navg_duration: ", results_by_county['duration'].mean(),"\nfrequency:", len(results_by_county),"\navg_customer_affected_mean: ",results_by_county['average_customers_out'].mean(),"\ntotal_customer_affected_mean_duration: ",results_by_county['weighted_customers_out'].mean(),"\ntotal_customer_affected_x_duration: ",results_by_county['weighted_customers_out'].sum())
            duration_diff = abs(results_by_county['duration'].mean() - answer_avg_duration)
            frequency_diff = abs(len(results_by_county) - answer_frequency)

            if duration_diff < best_avg_duration_diff:
                best_avg_duration_diff = duration_diff
                best_dur_mod = mod
                best_dur_pen = pen_val
                best_dur_change_points = change_points_by_county.to_dict()

            if frequency_diff < best_frequency_diff:
                best_frequency_diff = frequency_diff
                best_fre_mod = mod
                best_fre_pen = pen_val
                best_fre_change_points = change_points_by_county.to_dict()

        except Exception as e:
            print(f"Error processing model {mod} with penalty {pen_val}: {e}")
```

```python
#earlier codes
def detect_outages(data, model, penalty):
    if data.empty:
        return pd.DataFrame(), []

    # Reset index and ensure data is cleaned up
    data = data.reset_index(drop=True)
    data['CustomersOut'] = pd.to_numeric(data['CustomersOut'], errors='coerce').fillna(0)
    data['CustomersTracked'] = pd.to_numeric(data['CustomersTracked'], errors='coerce').fillna(1)
#     data['OutageRatio'] = data['CustomersOut'] / data['CustomersTracked']
#     data['OutageRatio'].replace([float('inf'), -float('inf'), pd.NA, pd.NaT], 0, inplace=True)

    segments = []
    current_segment_start = 0
    last_non_zero = None
    prev_zero_TF = False

    for i in range(len(data)):
        current_value = data['CustomersOut'][i]
        if current_value == 0:
            if last_non_zero is not None and prev_zero_TF == False:
                segments.append((current_segment_start, i + 1))
            if prev_zero_TF == False:
                current_segment_start = i + 1
            last_non_zero = None
            prev_zero_TF = True
        else:
            if last_non_zero is None:
                current_segment_start = i
            last_non_zero = current_value
            prev_zero_TF = False
            
    if last_non_zero is not None and current_segment_start < len(data):
        segments.append((current_segment_start, len(data)))

    all_segments = []
    outage_segments = []
    for start, end in segments:
        segment_data = data.iloc[start:end]
        for j in range(start + 1, end):
            if segment_data.at[j, 'CustomersOut'] < segment_data.at[j - 1, 'CustomersOut']:
                segment_data.at[j, 'CustomersOut'] = segment_data.at[j - 1, 'CustomersOut']
        if end - start > 2:
            segment_data['OutageRatio'] = segment_data['CustomersOut'] / segment_data['CustomersTracked']
            points = segment_data['OutageRatio'].values
            algo = rpt.Pelt(model=model, min_size=1, jump=1).fit(points)
            try:
                change_points = algo.predict(pen=penalty)
                actual_change_points = [start + cp for cp in change_points if cp != 0]  # Ensure no zero index changes
                all_segments.extend(actual_change_points)
                start = actual_change_points[-1] if actual_change_points else start
                if start < end:
                    all_segments.append(start)
            except Exception as e:
                print(f"Segmentation failed: {str(e)}")
                all_segments.append(start)
        else:
            all_segments.append(end)

    for start in all_segments:
        if start < len(data):  
            end = all_segments[all_segments.index(start) + 1] if all_segments.index(start) + 1 < len(all_segments) else len(data)
            segment = data.iloc[start:end]
            if not segment.empty:
                outage_segments.append({
                    'CountyFIPS': segment['CountyFIPS'].iloc[0],
                    'CityName': segment['CityName'].iloc[0],
                    'start_time': segment['RecordDateTime'].iloc[0],
                    'end_time': segment['RecordDateTime'].iloc[-1],
                    'average_customers_out': segment['CustomersOut'].mean(),
                    'duration': segment['RecordDateTime'].iloc[-1] - segment['RecordDateTime'].iloc[0],
                    'weighted_customers_out': segment['CustomersOut'].mean() * (segment['RecordDateTime'].iloc[-1] - segment['RecordDateTime'].iloc[0]).total_seconds(),
                    'segment_label': f"Outage_{start}"
                })

    return pd.DataFrame(outage_segments), all_segments
```
